***Merge, Join and Concatenate***


```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Python\practise\practise file resources\LOTR.csv")

df1
```

```python
df2 = pd.read_csv(r"C:\Python\practise\practise file resources\LOTR 2.csv")

df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2,how= 'inner' , on='FellowshipID')   # as we mergin on ID only so it shows name at dff
```

```python
df1.merge(df2,how= 'outer') 
```

```python
df1.merge(df2,how= 'left') 
```

```python
df1.merge(df2,how= 'cross') 
```

*** JOINS ***

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix="_left",rsuffix="_right")

df4
```

***Concat**


```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='inner')
```
