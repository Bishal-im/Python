```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Python\practise\practise file resources\Flavors.csv")

df
```

```python
group_by_frame = df.groupby('Base Flavor')          
```

```python
group_by_frame.mean(numeric_only=True)        #This ignores all string columns and averages only numeric ones (like Price, Calories, Rating, etc.).
```

```python
df.groupby('Base Flavor').count()
```

MIN , MAX 

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

------------------------------------------------------------------------------------------------------------------------------------------------------

***Combine aggrate***

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum']})
```

****MUltiple COlumn Grouping****

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby(['Base Flavor', 'Liked']).describe()
```
