***Automatic File Sorter in File Explorer***

```python
import os, shutil
```

```python
path = input("Enter the full path of the folder or file: ")
new_path = path + "\\"   # adds "/" at the end
print(new_path)

```

```python
file_name = os.listdir(new_path)  # list of files/folders in path

for i in range(len(file_name)):
    print(file_name[i])
```

```python
folder_names = ['csv files' ,'image files' ,'text files'];

for loop in range(len(folder_names)):
    
    if not os.path.exists(new_path + folder_names[loop] ):
        
        print(new_path + folder_names[loop] )
        os.makedirs((new_path + folder_names[loop]))

for file in file_name :
    if '.csv' in file and not os.path.exists(new_path +"csv files/" + file):
        shutil.move(new_path + file , new_path + "csv files/" + file)
    elif '.png' in file and not os.path.exists(new_path +"image files/" + file):
        shutil.move(new_path + file , new_path + "image files/" + file)
    elif '.txt' in file and not os.path.exists(new_path +"text files/" + file):
        shutil.move(new_path + file , new_path + "text files/" + file)
    else:
        print("There are files in this path that were not moved")
```
