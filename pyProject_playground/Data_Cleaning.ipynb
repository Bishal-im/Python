```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Python\practise\practise file resources\Customer Call List.xlsx")

df
```

```python
df1 = df.drop_duplicates()

df1
```

```python
df1 = df1.drop(columns="Not_Useful_Column")
```

```python
df1
```

```python
df1["Last_Name"] = df1['Last_Name'].str.strip("123._/")

df1
```

```python
df1["Phone_Number"] = df1["Phone_Number"].replace('[^a-zA-Z0-9]','',regex = True)

df
```

```python
df1["Phone_Number"] = df1["Phone_Number"].apply(lambda x: str(x) )

df1["Phone_Number"]                                             
```

```python

df1["Phone_Number"] = df1["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' +  x[6:11] )

df1["Phone_Number"]
```

```python
df1["Phone_Number"] = df1["Phone_Number"].str.replace("nan--",'')
```

```python
df1["Phone_Number"] = df1["Phone_Number"].str.replace("Na--",'')

df1["Phone_Number"]
```

```python
df1
```

```python
df1[['Street Address','State','Zip_Code']] = df1['Address'].str.split(',',n=2 , expand = True)
df1
```

```python
df1 = df1.drop(columns="Address")

df1
```

```python
df1['Paying Customer'] = df1['Paying Customer'].str.replace('Yes' , 'Y')

df1['Paying Customer'] = df1['Paying Customer'].str.replace('No' , 'N')

df1
```

```python
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('Yes' , 'Y')

df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('No' , 'N')

df1
```
