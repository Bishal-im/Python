```python
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python

# connect to website
URL = 'https://www.amazon.com/dp/B08KZ848BN/ref=sspa_dk_hqp_detail_aax_0?sp_csd=d2lkZ2V0TmFtZT1zcF9ocXBfc2hhcmVk&customId=B0752XJYNL&customizationToken=MC_Assembly_1%23B0752XJYNL&th=1&psc=1'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find(id='productTitle').get_text().strip()
price = soup2.find("span", {"class": "p13n-sc-price"}).get_text(strip=True)
Rating = soup2.find("span", {"class": "a-size-base a-color-base"}).get_text(strip=True)

print("Price :", price)
print("Title :",title)
print("Rating :", Rating)
```
